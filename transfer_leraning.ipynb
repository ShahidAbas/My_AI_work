```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.resnet import preprocess_input
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.applications.resnet import ResNet50
from tensorflow.keras.applications.resnet_v2 import ResNet152V2
from tensorflow.keras.applications.resnet_v2 import ResNet101V2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.densenet import DenseNet121
from tensorflow.keras.applications.efficientnet import EfficientNetB0
```

```python
vgg_base_model=VGG16(weights='imagenet', include_top=False, input_shape=(224,224,3))
```

```python
vgg19_base_model=VGG19(weights='imagenet', include_top=False, input_shape=(224,224,3))
```

```python
resnet50_base_model=ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
```

```python
resnet101_base_model=ResNet101V2(weights='imagenet', include_top=False, input_shape=(224,224,3))
```

```python
resnet152_base_model=ResNet152V2(weights='imagenet', include_top=False, input_shape=(224,224,3))
```

```python
denesnet121_base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224,244,3))
```

```python
efficientnetB0_base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224,244,3))
```

```python
dick={'vgg16':vgg_base_model,
       'vgg19':vgg19_base_model,
       'resnet50':resnet50_base_model,
       'resnet101':resnet101_base_model,
       'resnet152':resnet101_base_model,
       'densentnet121': denesnet121_base_model,
      'efficientnetB0': efficientnetB0_base_model
       }
```

```python
data_path = "/content/drive/MyDrive/Dataset and notebook/AIDER"
```

```python
datagen=ImageDataGenerator(preprocess_input,
                           zoom_range=0.2,
                           width_shift_range=0.2,
                           horizontal_flip=True,
                           validation_split=0.2


)
```

```python
train_data = datagen.flow_from_directory(
    data_path,
    target_size = (224,224),
    batch_size = 16,
    class_mode = 'categorical',
    subset = "training"
)
valid_data = datagen.flow_from_directory(
     data_path,
    target_size = (224,224),
    batch_size = 16,
    class_mode = 'categorical',
    subset = "validation"
)
```

```python
from typing import Any
for model_name, model in dick.items():
    print(f"loading {model_name}")
    print(f"Graph of model {model_name}")
    # if model_name='vgg16':
    #     continue
    base_model=model
    for layer in base_model.layers:
        layer.trainable=False
    x=base_model.output
    x=layers.Flatten()(x)
    x=layers.Dense(1024,activation='relu')(x)
    predictions=layers.Dense(5, activation='softmax')(x)
    proposed_model=keras.models.Model(inputs=base_model.input, outputs=predictions)
    #compile
    proposed_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    history=proposed_model.fit(train_data, epochs=20, validation_data=valid_data)

    # summarize history for accuracy
    print(f"Graph of model {model_name}")
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

    proposed_model.save(f'{model_name}.h5')
```
