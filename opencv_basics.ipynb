# function_for_displying_image

```python
def show_image(title="", image=None, size=5):
    h,w = image.shape[0], image.shape[1]
    ratio = w/h
    plt.figure(figsize=(ratio * size, size))
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image2 = cv.imread('../opencv_images/images/abraham.jpg')
```

```python
show_image(title="2nd_image", image=image2)
```

# bitwise operations

```python
image1 = np.zeros((300,300), np.uint8)
image12 = np.zeros((300,300), np.uint8)
cv.rectangle(image1, (50,50), (150,150), 255, -2)
cv.rectangle(image12, (80,80), (180,180), 255, -2)
Not=cv.bitwise_not(image1)
```

```python
show_image(title="empty_image1", image=image1, size=5)
show_image(title="empty_image12", image=image12, size=5)
show_image(title="Not_image", image=Not, size=5)
```

```python
And = cv.bitwise_and(image1, image12)
```

```python
show_image(title="AND", image=And, size=5)
```

```python
oR = cv.bitwise_or(image1, image12)
```

```python
show_image(title="OR", image=oR, size=5)
```

```python
xor = cv.bitwise_xor(image1, image12)
```

```python
show_image(title="XOR", image=xor, size=5)
```

```python
xor_not = cv.bitwise_not(xor)
```

```python
show_image(title="xor_not", image=xor_not, size=5)
```

# thresholding

# Erosion and dilation

```python
im10=cv.imread("../opencv_images/images/opencv.png")
```
